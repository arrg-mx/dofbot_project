# Objetivos

**Objetivo general:**


&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Colocar y ajustar correctamente la cámara Logitech C920 en una estructura fija que permita capturar imágenes y abarcar el espacio de trabajo donde se detectarán los marcadores AprilTags para el funcionamiento del sistema.


**Objetivos específicos**

1.  Diseñar e imprimir en 3D un soporte para la cámara Logitech C920 y colocar una estructura estable y ajustable utilizando perfiles de aluminio para montar la cámara.
2. Posicionar la cámara en el ángulo y altura adecuados para cubrir de manera eficiente el área de trabajo del robot.
3. Documentar el proceso de montaje y configuración de la cámara para su integración en el sistema general del proyecto.
# Introduccion 

En la actualidad, las cámaras juegan un papel crucial en la interacción de sistemas robóticos con su entorno, permitiendo identificar marcadores y procesar información visual. En este proyecto, se implementó una cámara Logitech C920 con el objetivo de integrarla en un sistema de detección de marcadores AprilTags. Para ello, fue necesario diseñar una estructura estable y funcional que permitiera fijar y ajustar la cámara en el ángulo y posición adecuados.

# Desarrollo 

Para la implementación de la cámara Logitech C920, se realizaron modificaciones en la estructura ubicada en el tercer piso del Centro de Innovación y Aprendizaje (CIA) utilizando los siguientes materiales y equipo:


**Materiales y equipo**

-  Cámara Logitech C920 
-  1 perfil de aluminio 40x40x65. 
-  1 perfil de aluminio 40x40x60. 
-  4 soportes de esquina en ángulo recto 28x35x35 mm. 
-  8 tornillos de ranura en T M8x20. 
-  8 tuercas de sujeción M8. 
-  1 tuerca en T M6. 
-  2 tornillos M6x15. 
-  Soporte para cámara Logitech C920 (impreso en 3D). 

Además, se elaboró un modelo digital del soporte en Autodesk Inventor archivo *Holder.ipt*, y se generó el archivo *Holder.stl* para la impresión 3D del soporte.


**Procedimiento**


1. Se ensamblaron los perfiles de aluminio en orientación perpendicular utilizando los soportes de esquina en ángulo recto, junto con los tornillos de ranura en T M8x20 y las tuercas M8, tal como se indica en la figura adjunta.


![image_0.png](Reporte_final_media/image_0.png)



2. La cámara Logitech C920 se fijó al soporte previamente fabricado mediante dos tornillos M6x15.


3. El tornillo M6 expuesto en el soporte se ajustó en la intersección de los perfiles utilizando una tuerca en T M6, asegurando el correcto posicionamiento y estabilidad de la cámara.


![image_1.png](Reporte_final_media/image_1.png)



**Resultado**


El ensamblaje final se muestra en la siguiente imagen, destacando la correcta integración de los materiales y el posicionamiento funcional de la cámara.


![image_2.png](Reporte_final_media/image_2.png)

# Objetivos

**Objetivos específicos**

1.  **Diseñar y fabricar un marcador físico para delimitar el uso de espacio del robot dentro de la mesa de trabajo.**
2. **Diseñar, fabricar y ensamblar una estructura tipo estantería para la interacción eficiente con el robot, así como integrar la estructura tipo estantería con el sistema robotizado.**
3. **Evaluar la efectividad del espacio de trabajo delimitado y la estantería en la operativa del robot.**
# Introduccion

El **proyecto de construcción del espacio de trabajo** tiene como objetivo optimizar el entorno en el que opera el robot **Dofbot**, asegurando una interacción precisa y eficiente con sus componentes. Para ello, se diseñó y fabricó un **marcador delimitador** que establece claramente el área de trabajo del robot, evitando posibles interferencias durante sus movimientos. Además, se desarrolló una **estructura tipo estantería** que proporciona soporte y organización a los equipos necesarios para la operación del robot, permitiendo un acceso fácil y seguro a los dispositivos periféricos. Ambas soluciones fueron diseñadas para mejorar la funcionalidad y la seguridad en la operación del Dofbot, garantizando un entorno controlado y adecuado para su interacción.

# Desarrollo

En el marco del proyecto titulado **"Construcción del Espacio de Trabajo 2"**, la tarea principal consistió en el diseño y construcción de dos elementos fundamentales para la correcta operación del robot **Dofbot** dentro de un espacio delimitado. Mi contribución se enfocó en la creación de un marcador delimitador para el área de trabajo, así como el diseño y fabricación de una estructura tipo estantería para facilitar la interacción del robot con sus componentes.


**Materiales y equipo**

-  **Software Autodesk Inventor 2025** 
-  **Madera MDF** 
-  **Cortadora láser** 
-  **Pegamento para madera** 

**1. Diseño y Construcción del Marcador del Espacio de Trabajo**


Para delimitar el área en la que el robot debe operar, se diseñó un **marcador físico** que actúa como frontera visible para el robot y los operadores. Este marcador fue creado en el software CAD **Autodesk Inventor**, permitiendo modelar las piezas con precisión según las dimensiones del área de trabajo del Dofbot.


![image_3.png](Reporte_final_media/image_3.png)


**2. Diseño y Construcción de la Estructura Tipo Estantería**


El siguiente paso fue el diseño de una **estructura tipo estantería**, destinada a organizar los componentes necesarios para la operación del Dofbot. El diseño de la estantería también se realizó en **Autodesk Inventor**, optimizando el espacio y asegurando su estabilidad y accesibilidad.


![image_4.png](Reporte_final_media/image_4.png)


Una vez completado el diseño, se utilizó **corte láser** para fabricar las piezas de la estantería en **MDF**, un material adecuado por su rigidez y facilidad para el ensamblaje. El ensamblaje de la estructura se realizó con pegamento para madera extra fuerte, asegurando que la estantería fuera estable y segura para los componentes que albergaría.


![image_5.png](Reporte_final_media/image_5.png)     ![image_6.png](Reporte_final_media/image_6.png)


![image_7.png](Reporte_final_media/image_7.png)


**3. Integración y Evaluación**


Una vez que el **marcador** y la **estructura tipo estantería** fueron fabricados y ensamblados, se procedió a la integración de ambos elementos en el espacio de trabajo del robot. Se posicionó el marcador en los puntos estratégicos del área de trabajo y la estantería en un lugar accesible pero que no interfiera con el movimiento del robot.


Las pruebas de funcionalidad confirmaron que ambos elementos cumplían su propósito, mejorando la operativa del robot y garantizando un entorno de trabajo seguro y eficiente.
