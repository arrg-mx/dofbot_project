# Marco Teórico

## DOFBOT: Brazo Robótico de 6 Grados de Libertad
El **DOFBOT** es un manipulador robótico serial que posee 6 grados de libertad (DOF, por sus siglas en inglés), lo que le permite realizar movimientos complejos en un espacio tridimensional. Sus componentes principales incluyen motores servo, una garra para la manipulación de objetos, y una estructura liviana que facilita su uso en proyectos de investigación y enseñanza.  

Los **grados de libertad** se refieren a la cantidad de movimientos independientes que un robot puede realizar. En el caso del DOFBOT, los DOF están distribuidos en:
- Base rotativa.
- Articulaciones intermedias para movimientos angulares.
- Una garra para la manipulación directa de objetos.

El DOFBOT es compatible con controladores como Arduino y placas basadas en ROS (Sistema Operativo Robótico), lo que permite la integración de algoritmos avanzados para la cooperación entre robots.

## Cinemática y Control en Robótica Serial
El DOFBOT se basa en principios de cinemática directa e inversa:
- **Cinemática directa:** Consiste en calcular la posición y orientación del efector final (la garra) en función de los ángulos de las articulaciones.
- **Cinemática inversa:** Es el cálculo inverso, donde se determinan los ángulos necesarios para que el efector final alcance una posición deseada. Esto es esencial para tareas como manipulación y colocación de objetos.

El control del DOFBOT incluye el uso de servomotores de precisión, programables para alcanzar ángulos específicos con alta exactitud. Además, su diseño permite realizar tareas como:
- Pruebas de movilidad y alcance.
- Manipulación de objetos en diferentes niveles de altura.

## Autodesk Inventor: Herramienta de Modelado 3D
El software Autodesk Inventor es una herramienta de diseño asistido por computadora (CAD) que permite crear modelos tridimensionales precisos, esenciales para el diseño de componentes robóticos y objetos manipulables. En este proyecto, se utiliza Inventor para:
- Diseñar objetos como cajas y tazas con dimensiones específicas.
- Generar archivos en formato STL para impresión 3D.
- Realizar análisis de simulación estructural para garantizar la funcionalidad de los objetos en las pruebas.

Las ventajas de Inventor incluyen:
- Herramientas avanzadas de modelado paramétrico, que permiten ajustar dimensiones fácilmente.
- Simulación de ensamblajes para verificar cómo interactúan los objetos diseñados con el entorno del robot.
- Generación de planos técnicos para fabricar componentes adicionales si es necesario.

## Espacios de Trabajo y Diseño de Muebles
En robótica, el espacio de trabajo se define como el volumen tridimensional dentro del cual el robot puede operar. Este espacio depende directamente de los grados de libertad del robot y de las dimensiones físicas de sus componentes. El diseño del mueble para el DOFBOT considera:
- Altura y anchura adecuadas para que el robot alcance cada nivel sin restricciones.
- Estabilidad estructural para soportar las fuerzas ejercidas durante las pruebas.
- Ergonomía en la disposición de los objetos para maximizar la eficiencia en la manipulación.

## Impresión 3D y Diseño de Objetos
La impresión 3D es un recurso clave para crear objetos que el DOFBOT manipulará. Este proceso incluye:
- **Modelado 3D:** Creación de diseños en Inventor con medidas específicas.
- **Formato STL:** Exportación de los modelos para ser procesados por una impresora 3D.
- **Materiales:** Uso de filamentos como PLA, que son ligeros y fáciles de imprimir.

Los objetos diseñados deben cumplir con criterios de funcionalidad, como ser lo suficientemente resistentes para soportar las fuerzas aplicadas por la garra del DOFBOT y tener formas que faciliten su agarre.

## Esquemas de Cooperación entre Robots
La cooperación entre robots implica la programación de múltiples manipuladores para trabajar de manera sincronizada en una tarea. Esto requiere:
- **Algoritmos de control colaborativo:** Coordinar los movimientos de varios robots para realizar tareas complejas, como el ensamblaje de componentes.
- **Pruebas de sincronización:** Evaluar la capacidad de los robots para compartir un espacio de trabajo y realizar movimientos complementarios.
- **Comunicación remota:** Uso de protocolos de comunicación para controlar los robots desde un sistema centralizado.

Los esquemas de cooperación aumentan la eficiencia de los sistemas robóticos, especialmente en aplicaciones donde un solo robot no puede completar todas las tareas por sí mismo.

## Pruebas de Movilidad y Manipulación
El análisis de movilidad del DOFBOT se centra en:
- Evaluar su capacidad para alcanzar diferentes posiciones dentro del mueble.
- Determinar los límites físicos del robot, como el alcance máximo de la garra y los ángulos extremos de las articulaciones.
- Verificar la precisión con la que el robot puede manipular y colocar objetos en espacios definidos.

Estas pruebas son fundamentales para programar los esquemas de cooperación y garantizar que los robots puedan realizar tareas sin interferencias ni errores.
